```python
#https://github.com/awesomebytes/python-control-code/blob/master/src/matlab.py
%matplotlib inline
import pylab as plt
from control import tf
import control
import numpy as np
import matplotlib
matplotlib.rcParams.update({'font.size': 18})
matplotlib.rcParams['figure.figsize'] = (11.0, 6.)
```

```python
#MATLAB Program 2–1
num1 = [10]
den1 = [1, 2, 10]
num2 = [5]
den2 = [1,5]
valor1 = tf(num1,den1)
valor2 = tf(num2,den2)
G1 = control.series(valor1, valor2)
G2 = control.parallel(valor1,valor2)
G3 = control.feedback(valor1,valor2)
print(G1)
print(G2)
print(G3)
```

```python
#MATLAB Program 2–4
num = [2, 1, 1, 2]
den = [1, 4, 5, 2]
sys = control.tf2ss(num,den)
print (sys)
```

```python
#MATLAB Program 5–1
#A = [(-1, -1), (6.5, 0)]
#B = [(1,1), (1,0)]
#C = [(1,0), (0,1)]
#D = [(0,0), (0,0)]
den = [1, 1, 6.5]
y1u1 = tf([1,-1],den)
y1u2 = tf([1,0], den)
y2u1 = tf([1,7.5],den)
y2u2 = tf([6.5], den)
t1, out1 = control.step_response(y1u1)
t2, out2 = control.step_response(y1u2)
t3, out3 = control.step_response(y2u1)
t4, out4 = control.step_response(y2u2)

plt.plot(t1,out1,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.grid(True)
plt.plot(t2,out2,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.grid(True)
plt.plot(t3,out3,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.grid(True)
plt.plot(t4,out4,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.grid(True)
plt.legend(["y1u1","y1u2","y2u1","y2u2"])
```

```python
#MATLAB Program 5–2

plt.plot(t1,out1,lw = 3)
plt.plot(t3,out3,lw = 3)
plt.grid(True)
plt.title ('Step-Response Plots: Input = u1 (u2 = 0)')
plt.text(3.4, -0.06,'Y1')
plt.text(3.4, 1.4,'Y2')
```

```python
#MATLAB Program 5–2 Continuació

plt.plot(t2,out2,lw = 3)
plt.plot(t4,out4,lw = 3)
plt.grid(True)
plt.title ('Step-Response Plots: Input = u2 (u1 = 0)')
plt.text(3,0.14,'Y1')
plt.text(2.8,1.1,'Y2')
```

```python
#MATLAB Program 5–6
t = np.linspace(0,0.20,10)
zeta = [0, 0.2, 0.4, 0.6, 0.8, 1]
n = 0
while n<6:
    num = [1]
    valor = zeta[n] * 2
    den = [1, valor, 1]
    trans = tf(num,den)
    t, out = control.step_response(trans)
    n = n + 1
    plt.plot(t,out)

plt.grid(True)    
plt.title('Plot of Unit-Step Response Curves with \omega_n = 1 and \zeta = 0, 0.2, 0.4, 0.6, 0.8, 1')
plt.xlabel('t (sec)')
plt.ylabel('Response')
plt.text(4.1,1.86,'\zeta = 0')
plt.text(3.5,1.5,'0.2')
plt.text(3.5,1.24,'0.4')
plt.text(3.5,1.08,'0.6')
plt.text(3.5,0.95,'0.8')
plt.text(3.5,0.86,'1.0')

```
