```python
%matplotlib inline  
import itertools
import numpy as np
import matplotlib
from matplotlib import pyplot as plt 

from sklearn.mixture import GaussianMixture as GMM
from igmm import IGMM
import time
```

```python
def plot_gmm_projection(model_tmp,axes):
    model = IGMM(min_components=2, max_step_components=5, max_components=5)
    model.covariances_ = model_tmp.covariances_
    model.means_ = model_tmp.means_
    model.weights_ = model_tmp.weights_
    model.n_components = model_tmp.n_components
    return model.plot_gmm_projection(0,1, axes=axes)
```

```python
np.random.seed(0)
n_samples = 100 # Samples per component
model = GMM(n_components=5)
alpha = 0.05
```

```python
matplotlib.rcParams['figure.figsize'] = (8.0, 8.0)
matplotlib.rcParams.update({'font.size': 18})
title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'size':'30'}
```

```python
# Batch 1
C1 = np.array([[0., -0.1], [1.7, .4]])
X1 = np.r_[np.dot(np.random.randn(n_samples, 2), C1),
          .7 * np.random.randn(n_samples, 2) + np.array([-6, 3])]
model.fit(X1)
fig1, ax1 = plt.subplots(1,1)
ax1 = plot_gmm_projection(model, axes=ax1)
plt.scatter(X1[:,0],X1[:,1],c='k', marker='.')
plt.xlabel('$x$',**axis_font)
plt.ylabel('$y$',**axis_font)
```

```python
matplotlib.rcParams['figure.figsize'] = (18.0, 7.0)

# Batch 2
C2 = np.array([[.8, .2], [0.1, -.2]])
C3 = np.array([[0.1, -0.1], [1., .8]])
C4 = np.array([[-0.8, 1.], [-0.1, -0.1]])
fig2, ax2 = plt.subplots(1,3)
ax2[0] = plot_gmm_projection(model, axes=ax2[0])
plt.title('$M_{t-2}$', y=1.01)
plt.ylabel('$y$',**axis_font)
plt.scatter(X1[:,0],X1[:,1],c='k', marker='.')

X2 = np.r_[np.dot(np.random.randn(n_samples, 2), 0.5 * C1) + np.array([-5, 4]),
           np.dot(np.random.randn(n_samples, 2), C2) + np.array([1, 1]),
           np.dot(np.random.randn(n_samples, 2), 0.5 * C3) + np.array([4, 3]),
         np.dot(np.random.randn(n_samples, 2), 0.5 * C4) + np.array([-2, 1])]

gen_samples = int(np.ceil((1.-alpha)*n_samples*4/alpha))
X1_gen, foo = model.sample(n_samples=gen_samples)
plt.sca(ax2[1])
plt.title("$\mathbf{Z}=\mathbf{Z}_{t-1}'+\mathbf{Z}_{t-1}$", y=1.01)

plt.scatter(X1_gen[:,0],X1_gen[:,1],c='r', marker='.',edgecolors='r')
plt.scatter(X2[:,0],X2[:,1],c='k', marker='o')
plt.legend(["$\mathbf{Z}_{t-1}'$", "$\mathbf{Z}_{t-1}$"],loc=4)

model.fit(np.concatenate((X2,X1_gen),axis=0))
ax2[2] = plot_gmm_projection(model, axes=ax2[2])
plt.title('$M_{t-1}$', y=1.01)
for i in range(3):
    plt.sca(ax2[i])
    ax2[i].set_ylim([-2,7])
    ax2[i].set_xlim([-10,8])
    plt.xlabel('$x$',**axis_font)
```

```python
# Batch  3
fig3, ax3 = plt.subplots(1,3)
ax3[0] = plot_gmm_projection(model, axes=ax3[0])
plt.ylabel('$y$',**axis_font)
plt.title('$M_{t-1}$', y=1.01)

X3 = np.r_[np.dot(np.random.randn(n_samples, 2), 0.5 * C1),
           .1 * np.random.randn(n_samples, 2) + np.array([-5, 4]),
           .1 * np.random.randn(n_samples, 2) + np.array([-2, 1]),
           .4 * np.random.randn(n_samples, 2) + np.array([-6, 0]),
           .4 * np.random.randn(n_samples, 2) + np.array([4, 5])]
gen_samples = int(np.ceil((1.-alpha)*n_samples*5/alpha))
X12_gen, foo = model.sample(n_samples=gen_samples)
plt.sca(ax3[1])
plt.title("$\mathbf{Z}=\mathbf{Z}_{t}+\mathbf{Z}_{t}'$", y=1.01)
plt.scatter(X12_gen[:,0],X12_gen[:,1],c='r', marker='.',edgecolors='r')
plt.scatter(X3[:,0],X3[:,1],c='k', marker='.')
plt.legend(["$\mathbf{Z}_{t}'$", "$\mathbf{Z}_{t}$"],loc=4)

model.fit(np.concatenate((X3,X12_gen),axis=0))
ax3[2] = plot_gmm_projection(model, axes=ax3[2])
plt.title('$M_{t}$', y=1.01)
for i in range(3):
    plt.sca(ax3[i])
    ax3[i].set_ylim([-2,7])
    ax3[i].set_xlim([-10,8])
    plt.xlabel('$x$',**axis_font)
```
