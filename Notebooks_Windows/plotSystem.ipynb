```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
from parabola import ParabolicRegion as System
```

```python
system = System()
```

```python
matplotlib.rcParams['figure.figsize'] = (6.0, 9.0)
matplotlib.rcParams.update({'font.size': 18})

title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'size':'30'}

fig, ax = plt.subplots(1,1) 
plt.hold(True)
system.drawSystem(fig, ax)
ax.set_xlim([0,6])
ax.set_ylim([0,9])
plt.xlabel('$y_1$',**axis_font)
plt.ylabel('$y_2$',**axis_font)
plt.legend(['Constraints'],loc='upper center',fontsize=15)
```
